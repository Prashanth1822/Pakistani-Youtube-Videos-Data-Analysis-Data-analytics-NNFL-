```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/Pak_videos.csv')  # replace 'your_dataset.csv' with the path to your file
```

```python
# Group by 'category' and sum 'likes' and 'comments' for each category
category_summary = df.groupby('category')[['likes', 'comments_count']].sum()

# Display the summary
print(category_summary)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data for plotting
categories = category_summary.index
likes = category_summary['likes']
comments = category_summary['comments_count']
x = np.arange(len(categories))  # label locations for each category
width = 0.35  # width of the bars

# Create the grouped bar chart
fig, ax = plt.subplots(figsize=(12, 7))
bars1 = ax.bar(x - width/2, likes, width, label='Likes', color='skyblue')
bars2 = ax.bar(x + width/2, comments, width, label='Comments_count', color='salmon')

# Adding labels, title, and legend
ax.set_xlabel('Category')
ax.set_ylabel('Number of Interactions')
ax.set_title('Total Likes and Comments by Category')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('/content/Pak_videos.csv')

# Check the first few rows of the dataset
print(df.head())
```

```python
print("Available columns:", df.columns)
```

```python
# Prepare data for the pie chart
total_positive = df['positive_videos'].sum()
total_negative = df['negative_videos'].sum()
labels = ['Positive Videos', 'Negative Videos']
colors = ['blue', 'red']

# Create the pie chart
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie([total_positive, total_negative], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'alpha': 0.7})

# Customize the plot
ax.set_title('Distribution of Positive and Negative Videos', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def duration_to_seconds(duration):
    try:
            time = pd.to_timedelta(duration)
            return time.total_seconds()
    except:
            return None

df = pd.read_csv('/content/Pak_videos.csv')  # Load your data
df['duration_seconds'] = df['duration'].apply(duration_to_seconds)
```

```python
category_duration = df.groupby('category')['duration_seconds'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(category_duration['category'], category_duration['duration_seconds'], color='skyblue')
plt.title('Average Video Duration by Category')
plt.xlabel('Category')
plt.ylabel('Average Duration (seconds)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
